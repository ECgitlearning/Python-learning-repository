```python
# if -> run the code if a condition is true
# elif -> will run the code if it is true AND previous conditions were not true
# else -> runs if none of the previous conditions were true

names_list = ['Will', 'Clare', 'Marcel', 'Alistair', 'Joe']

for n in names_list:
    if n == 'Marcel':
        print('Woof!')
    elif n == 'Will':
        print(f"Hi, I'm Will!")
    else:
        print(f'Hello, my name is {n}')
    if n == 'Alistair':
        print("I'm Will's boss.")
```

```python
# Loops inside Loops

 
evens = list(range(2,13,2))
odds = list(range(1,12,2))

multiple_list = []
for even in evens:
    for odd in odds: # nested loop 
        result = even * odd
        stmt = f'{even} times {odd} = {result}'
        multiple_list.append(stmt)
print(multiple_list)
 

```

```python
# % modulus 

print(80 % 9 == 0)

# fizzbang
# take all the numbers from 1-100, 
# print them, BUT if the number is divisible by 3, print fizz,
# if it's divisible by 5 print bang
# BUT if it's divisible by 3 & 5 print fizzbang

num_list = list(range(1,101))

for num in num_list:
    if (num % 3 == 0) & (num % 5 == 0):
        print('fizzbang') 
    elif num % 3 == 0:
        print('fizz')
    elif num % 5 == 0:
        print('bang')
    else:
        print(num)
```
